```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
```

```python
#load real dataset
data=load_breast_cancer()
X=data.data
Y=data.target

print("Classes:",data.target_names)
#malignant, benign

#split data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=30)
```

### Logistic Regression

```python
lr= LogisticRegression(max_iter=5000)
lr.fit(X_train,Y_train)
pred_lr=lr.predict(X_test)
print("Logistic TestAccuracy:",accuracy_score(Y_test,pred_lr))
pred_lr_train=lr.predict(X_train)
print("Logistic TrainAccuracy:",accuracy_score(Y_train,pred_lr_train))
```

### Decision Tree

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)
pred_dt = dt.predict(X_test)
print("Decision Tree TestAccuracy:", accuracy_score(Y_test, pred_dt))
pred_dt_train = dt.predict(X_train)
print("Decision Tree TrainAccuracy:", accuracy_score(Y_train, pred_dt_train))
```

### kNN

```python
knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
pred_knn = knn.predict(X_test)
print("KNN TestAccuracy:", accuracy_score(Y_test, pred_knn))
pred_knn_train = knn.predict(X_train)
print("KNN TrainAccuracy:", accuracy_score(Y_train, pred_knn_train))
```

### Naive Bayes

```python
nb = GaussianNB()
nb.fit(X_train, Y_train)
pred_nb = nb.predict(X_test)
print("Naive Bayes TestAccuracy:", accuracy_score(Y_test, pred_nb))
pred_nb_train = nb.predict(X_train)
print("Naive Bayes TrainAccuracy:", accuracy_score(Y_train, pred_nb_train))
```

### Neural Network

```python
nn = MLPClassifier(max_iter=1000)
nn.fit(X_train, Y_train)
pred_nn = nn.predict(X_test)
print("Neural Network TestAccuracy:", accuracy_score(Y_test, pred_nn))
pred_nn_train = nn.predict(X_train)
print("Neural Network TrainAccuracy:", accuracy_score(Y_train, pred_nn_train))
```

### Classification Performance Metrics

```python
precision_lr = precision_score(Y_test, pred_lr)
recall = recall_score(Y_test, pred_lr)
f1 = f1_score(Y_test, pred_lr)
roc_auc_lr = roc_auc_score(Y_test, pred_lr)
print("Logistic Regression Metrics:")
print("Precision:", precision_lr)
print("Recall:", recall)
print("F1 Score:", f1)
print("ROC AUC Score:", roc_auc_lr)
```

```python
precision_dt = precision_score(Y_test, pred_dt)
recall = recall_score(Y_test, pred_lr)
f1 = f1_score(Y_test, pred_dt)
roc_auc_dt = roc_auc_score(Y_test, pred_dt)
print("Decision Tree Metrics:")
print("Precision:", precision_dt)
print("Recall:", recall)
print("F1 Score:", f1)
print("ROC AUC Score:", roc_auc_dt)
```

```python
precision_knn = precision_score(Y_test, pred_knn)
recall = recall_score(Y_test, pred_knn)
f1 = f1_score(Y_test, pred_knn)
roc_auc_knn = roc_auc_score(Y_test, pred_knn)
print("KNN Metrics:")
print("Precision:", precision_knn)
print("Recall:", recall)
print("F1 Score:", f1)
print("ROC AUC Score:", roc_auc_knn)
```

```python
precision_nb = precision_score(Y_test, pred_nb)
recall = recall_score(Y_test, pred_nb)
f1 = f1_score(Y_test, pred_nb)
roc_auc_nb = roc_auc_score(Y_test, pred_nb)
print("Naive Bayes Metrics:")
print("Precision:", precision_nb)
print("Recall:", recall)
print("F1 Score:", f1)
print("ROC AUC Score:", roc_auc_nb)
```

```python
precision_nn = precision_score(Y_test, pred_nn)
recall = recall_score(Y_test, pred_nn)
f1 = f1_score(Y_test, pred_nn)
roc_auc_nn = roc_auc_score(Y_test, pred_nn)
print("Neural Network Metrics:")
print("Precision:", precision_nn)
print("Recall:", recall)
print("F1 Score:", f1)
print("ROC AUC Score:", roc_auc_nn)
```

```python
#Example of confusion matrix for Decision Tree
cm_dt = confusion_matrix(Y_test, pred_dt)
print("Confusion Matrix for Decision Tree:")
print(cm_dt)
TN2, FP2, FN2, TP2 = cm_dt.ravel()
type1_dt = FP2 / (FP2 + TN2)
type2_dt = FN2 / (TP2 + FN2)

print("FP (Type-I):", FP2)
print("FN (Type-II):", FN2)
print("Decision Tree Type-I Error Rate in %:", type1_dt*100)
print("Decision Tree Type-II Error Rate in %:", type2_dt*100)
```

```python
confusion_matrix_lr = confusion_matrix(Y_test, pred_lr)
print("\n Logistic Regression Confusion Matrix:")
print(confusion_matrix_lr)

TN3, FP3, FN3, TP3 = confusion_matrix_lr.ravel()
type_1_lr = FP3 / (FP3 + TN3)
type_2_lr = FN3 / (TP3 + FN3)
print("Logistic Regression Type-I Error Rate in %:", type_1_lr*100)
print("Logistic Regression Type-II Error Rate in %:", type_2_lr*100)
```

```python
confusion_matrix_knn = confusion_matrix(Y_test, pred_knn)
print("\n KNN Confusion Matrix:")
print(confusion_matrix_knn)

TN3, FP3, FN3, TP3 = confusion_matrix_knn.ravel()
type_1_knn = FP3 / (FP3 + TN3)
type_2_knn = FN3 / (TP3 + FN3)
print("KNN Type-I Error Rate in %:", type_1_knn*100)
print("KNN Type-II Error Rate in %:", type_2_knn*100)
```

```python
confusion_matrix_nb = confusion_matrix(Y_test, pred_nb)
print("\n Naive Bayes Confusion Matrix:")
print(confusion_matrix_nb)

TN3, FP3, FN3, TP3 = confusion_matrix_nb.ravel()
type_1_nb = FP3 / (FP3 + TN3)
type_2_nb = FN3 / (TP3 + FN3)
print("Naive Bayes Type-I Error Rate in %:", type_1_nb*100)
print("Naive Bayes Type-II Error Rate in %:", type_2_nb*100)
```

```python
confusion_matrix_nn = confusion_matrix(Y_test, pred_nn)
print("\n Neural Network Confusion Matrix:")
print(confusion_matrix_nn)

TN3, FP3, FN3, TP3 = confusion_matrix_nn.ravel()
type_1_nn = FP3 / (FP3 + TN3)
type_2_nn = FN3 / (TP3 + FN3)
print("Neural Network Type-I Error Rate in %:", type_1_nn*100)
print("Neural Network Type-II Error Rate in %:", type_2_nn*100)
```
